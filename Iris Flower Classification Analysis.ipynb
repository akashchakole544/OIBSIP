```python
ax = sns.countplot(x='Species',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x='SepalLengthCm',data=df.head(20))
plt.show()
```

```python
ax = sns.histplot(x='SepalLengthCm',data=df,hue='Species')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.histplot(x='SepalWidthCm',data=df,hue='Species')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.histplot(x='PetalLengthCm',data=df,hue='Species')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=df,hue='Species',palette="deep")
plt.show()
```

```python
sns.lmplot(x='SepalLengthCm',y='SepalWidthCm',data=df,hue='Species')
plt.show()
```

```python
sns.lineplot(x='PetalLengthCm',y='PetalWidthCm',hue='Species',data=df)
plt.show()
```

```python
#we can see that iris setosa has values of petalwidth & petallength between 1 to 2
#iris versicolor has values between 3 to 5 for petalwidth & petallength
#iris-virginica has betweenn 4 to 7 for petalwidth & petallength
```

```python
sns.kdeplot(x='PetalWidthCm',y='SepalWidthCm',hue='Species',data=df,fill=True,)
plt.show()
```
